```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/rmd-davis/ML-intro/main/data/housing_data.csv')
```

### Basic EDA using built-in Pandas functions

```python
# Look at the first 5 rows of data
data.head(5)
```

```python
# Look at the columns
data.info()
```

```python
# Look at basic statistics for numerical data
data.describe().transpose()
```

### Automatic EDA using Pandas Profiler

```python
profile = ProfileReport(data, title='Housing Data', html={'style':{'full_width':True}})
profile.to_file(output_file = 'Housing_Data_EDA.html')
#profile.to_notebook_iframe()
```
